```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2D Random walk

We will concider three kinds of 2D random talk.
* $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.
* $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

For all kinds of the 2D random walk, you should do the following:
* Plot a single trajectory.
* Plot the distribution of the end point for an ensemble of trajectories.
* Plot the histogram of the end point for an ensemble of trajectories.

```python
def traj(step,dx):
    x=np.array([])
    for j in range(step):
        x=np.append(x,np.sum(dx[0:j]))
    return x
def distance(x,y):
    d=np.sqrt(x**2+y**2)
    return d

def hist_uni(walker,step):
    S=np.array([])
    for i in range(walker):
        dx=np.random.random((step,))-(1/2)
        dy=np.random.random((step,))-(1/2)
        d=distance(np.sum(dx),np.sum(dy))
        S=np.append(S,d)
    plt.hist(S,bins=50)

def disc(x):
    if x>=(1/2):
        return 1
    else:
        return -1
def hist_disc(walker,step):
    S=np.array([])
    for i in range(walker):
        dx=np.array([])
        dy=np.array([])
        r1=np.random.random((step,))
        r2=np.random.random((step,))
        for b in r1:
            dx=np.append(dx,disc(b))
        for b in r2:
            dy=np.append(dx,disc(b))
        d=distance(np.sum(dx),np.sum(dy))
        S=np.append(S,d)
    plt.hist(S,bins=50)
def hist_ang(walker,step):
    S=np.array([])
    for i in range(walker):
        dx=2*np.pi*np.random.random((step,))
        dy=2*np.pi*np.random.random((step,))-(1/2)
        d=distance(np.sum(dx),np.sum(dy))
        S=np.append(S,d)
    plt.hist(S,bins=50)
step=1000
walker=1000
```

```python
#unif 2-D traj 
dx=np.random.random((step,))-(1/2)
dy=np.random.random((step,))-(1/2)
x=traj(step,dx)
y=traj(step,dy)

plt.plot(x,y)
```

```python
#unif 2-D  hist
hist_uni(walker,step)
```

```python
#unif 2-D distribution
a=100
x=np.linspace(0,40)
plt.plot(x,2*x*(1/a)/(np.exp((x**2)/a)),'r')
```

```python
#2-D disc traj 
dx=np.array([])
dy=np.array([])
r1=np.random.random((step,))
r2=np.random.random((step,))
for b in r1:
    dx=np.append(dx,disc(b))
for b in r2:
    dy=np.append(dx,disc(b))
x=traj(step,dx)
y=traj(step,dy)
plt.plot(x,y)
```

```python
#2-D disc hist

hist_disc(walker,step)
```

```python
#2-D uniform on angle
dax=2*np.pi*np.random.random((step,))
day=2*np.pi*np.random.random((step,))
dx=np.cos(dax)
dy=np.sin(day)
x=traj(step,dx)
y=traj(step,dy)
plt.plot(x,y)
```

```python
#2-D uniform on angle hist
hist_ang(walker,step)
```
